```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
```

## Wczytywanie danych

```python
column_names = [
    'Class',
    'Alcohol',
    'Malic acid',
    'Ash',
    'Alcalinity of ash',
    'Magnesium',
    'Total phenols',
    'Flavanoids',
    'Nonflavanoid phenols',
    'Proanthocyanins',
    'Color intensity',
    'Hue',
    'OD280/OD315 of diluted wines',
    'Proline'
]

data = pd.read_csv('Dane/wine.data')
data.columns = column_names
display(data.head(10))
```

## Podział na zbiór treningowy i testowy

```python
train_data, test_data = train_test_split(data, test_size=0.3, random_state=42)

X_train = train_data.drop('Class', axis=1)
Y_train = train_data['Class']

X_test = test_data.drop('Class', axis=1)
Y_test = test_data['Class']
```

## Normalizacja danych

```python
scaler = StandardScaler()

scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

### Normalizuje się po to aby:
- dane miały porównywalną skalę
- niektóre algorytmy są wrażliwe na skalę (np. K-NN)
### Wpływ na algorytmy
- **k-NN** - musi być normalizacja, gdyż algorytm działa na podstawie odległości między punktami
- **Random Forest** - nie jest wymagana normalizacja bo drzewa decyzyjne porównują wartość do liczby niezależnie od skali

## Trening dla algorytmów KNeighborsClassifier oraz RandomForestClassifier

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, Y_train)

rfc = RandomForestClassifier()
rfc.fit(X_train, Y_train)
```

## Klasyfikacja

```python
y_pred_knn = knn.predict(X_test)
y_pred_rfc = rfc.predict(X_test)
```

## Metryki
- `Accuracy` - stosunek liczby poprawnych klasyfikacji do wszystkich przewidzianych, najlepiej działa, gdy klasy są w miarę zrównoważone liczebnie
- `Precision` - stosunek liczby poprawnych klasyfikacji w obrębie klasy do wszystkich przewidzianych danej klasy, najlepiej działa do sprawdzania fałszywych pozytywów
- `Recall` - ile z rzeczywistych przypadków danej klasy model wykrył
    - przydatne aby nie przeoczyć fałszywych negatywów
- `F-measure` - średnia harmoniczna `precision` i `recall`
    - dobry wskaźnik ogólnej jakości modelu, szczególnie przy niezrównoważonych klasach
- `Macierz konfuzji` - pokazuje, które klasy są najczęściej ze sobą mylone
    - Dobra do szybkiego wglądu w błędy modelu
- `Classification report` - zestawienie wszystkich powyższych metryk w tabeli, osobno dla każdej klasy

## Analiza predykcji poszczególnych modeli

```python
print("KNeighborsClassifier")
print("Accuracy = ", accuracy_score(Y_test, y_pred_knn))
print("Classification Report:\n", classification_report(Y_test, y_pred_knn))
print("Confusion Matrix:\n", confusion_matrix(Y_test, y_pred_knn))
```

```python
print("\nRandomForestClassifier")
print("Accuracy = ", accuracy_score(Y_test, y_pred_rfc))
print("Classification Report:\n", classification_report(Y_test, y_pred_rfc))
print("Confusion Matrix:\n", confusion_matrix(Y_test, y_pred_rfc))
```

## Interpretacja analizy
> Oba modele osiągnęły bardzo zbliżone wyniki. Analiza macierzy konfuzji pokazuje, że oba klasyfikatory miały trudności jedynie z poprawnym rozpoznaniem drugiego rodzaju wina. Klasyfikator oparty na drzewie decyzyjnym poradził sobie z tym nieco lepiej. Możliwe, że różnica ta wynika z różnych rozmiarów modeli, tzn. po zwiększeniu liczby sąsiadów w KNeighbors oba modele osiągnęły taką samą skuteczność.
